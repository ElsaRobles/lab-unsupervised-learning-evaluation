# Unsupervised Learning Model Evaluation Lab

Complete the exercises below to solidify your knowledge and understanding of unsupervised learning model evaluation.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
```

```python
from sklearn import datasets

data = datasets.load_wine()

X = pd.DataFrame(data["data"], columns=data["feature_names"])
y = pd.Series(data["target"])
```

```python
print(X)
```

## 1. Train a KMeans clustering model on the data set using 8 clusters and compute the silhouette score for the model.

```python
kmeans_model_8= KMeans(n_clusters=8)
kmeans_model_8.fit(X)
```

```python
score_8= silhouette_score(X,kmeans_model_8.labels_, metric='euclidean')
print(score_8)
```

## 2. Train a KMeans clustering model on the data set using 5 clusters and compute the silhouette score for the model.

```python
kmeans_model_5 = KMeans(n_clusters=5)
kmeans_model_5.fit(X, y)
```

```python
score_5 = silhouette_score(X,kmeans_model_5.labels_, metric='euclidean')
print(score_5)
```

```python
#The second model with less clusters (5, instead of 8) has a higher silhouette score
```

## 3. Train a KMeans clustering model on the data set using 3 clusters and compute the silhouette score for the model.

```python
kmeans_model_3 = KMeans(n_clusters=3)
kmeans_model_3.fit(X, y)
```

```python
score_3 = silhouette_score(X,kmeans_model_3.labels_, metric='euclidean')
print(score_3)
```

```python
#Finally, the 3rd one with 3 clusters, has a higher silhouette score. 
```

## 4. Use elbow curve visualizations to see if you can determine the best number of clusters to use.

The Yellowbrick library has 3 metrics that you can plot using the `metric` parameter:

- **distortion**: mean sum of squared distances to centers
- **silhouette**: mean ratio of intra-cluster and nearest-cluster distance
- **calinski_harabaz**: ratio of within to between cluster dispersion

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans

model = KMeans()
visualizer = KElbowVisualizer(model, k=(4,12))
visualizer.fit(X)
visualizer.poof();
```

```python
kmeans_model_7 = KMeans(n_clusters=7)
kmeans_model_7.fit(X, y)
```

```python
score_7 = silhouette_score(X,kmeans_model_7.labels_, metric='euclidean')
print(score_7)
```

```python
#Doubt: Why the if the best number of clusters acording to the Distorition Elbow Sore is 7, but the silhouette score with 3 clusters is higher? 
```

## 5. Try performing the same elbow tests with an AgglomerativeClustering model and compare the results you get to the KMeans results.

```python
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import AgglomerativeClustering

visualizer = KElbowVisualizer(AgglomerativeClustering(), k=(2,12), metric='distortion')
visualizer.fit(X)
visualizer.poof();
```

## 6. Create and plot a scatter matrix showing how the clusters are grouped across all the different combinations of variables in the data.

Use the model and number of clusters that returned the best result above.

```python
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

#According to the previus elbow curve visualizer, our optimum number of clusters is 4.
```

```python
X, _ = make_blobs(n_samples=1500, centers=4, cluster_std=1.5)

kmeans = KMeans(n_clusters=4)
kmeans.fit(X)
y_pred = kmeans.predict(X)

plt.figure(figsize=(12,8))
plt.scatter(X[:, 0], X[:, 1], c=y_pred)

centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=50, alpha=0.75)

plt.show();
```

## 7. Apply a PCA transform and plot the first two principle components with the plot point colors determined by cluster.

```python
from sklearn.decomposition import PCA

pca = PCA() #Initialize the PCA
pca.fit(X)
pca_X = pca.transform(X)
```

```python
fig = plt.figure(figsize=(12, 8))
plt.scatter(pca_X[:,0], pca_X[:,1], cmap="plasma")

plt.show();

#I was not able to plot the point colors by cluster. 
```

## 8. Generate a series of t-SNE plots showing the clusters at a variety of perplexities.

```python
from sklearn.manifold import TSNE

fig, ax = plt.subplots(1, 4, figsize=(15, 7))

for i, perp in enumerate([5, 30, 50, 100]):
    tsne = TSNE(perplexity=perp)
    x_embedded = tsne.fit_transform(X)
    ax[i].scatter(x_embedded[:, 0], x_embedded[:, 1], c=y)
    ax[i].set_title(f"Perplexity = {perp}")
    
plt.show();
```
